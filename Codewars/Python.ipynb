```python
'''
  Contando número de vogais em um string
'''

def get_count(sentence):
  vogais = ['a','e','i','o','u']
  count = 0
  for char in sentence.lower():
    if char in vogais:
      count = count + 1
  return count


sentence = 'aeiasdvsdou'
get_count(sentence)
```

```python
'''
  Trocando a letra do texto pelo número da posição no alfabeto
'''

def alphabet_position(text):
  alphabet = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6, 'g': 7, 'h': 8, 'i': 9, 'j': 10, 'k': 11, 'l': 12, 'm': 13, 'n': 14, 'o': 15, 'p': 16, 'q': 17, 'r': 18, 's': 19, 't': 20, 'u': 21, 'v': 22, 'w': 23, 'x': 24, 'y': 25, 'z': 26}
  alphabetList = []

  for char in text.lower():
    if char in alphabet.keys():
      alphabetList.append("{}".format(alphabet.setdefault(char)))

  return ' '.join(alphabetList)


text = "The sunset sets at twelve o' clock."
alphabet_position(text)
```

```python
'''
  Filtrar uma lista de string e caso a string tenha 4 letras , passar para outra lista
'''
def friend(x):
  friend = []
  for i in range(len(x)):
    if (len(x[i]) == 4):
      friend.append(x[i])
  return friend
  
x = ["Ryan", "Kieran", "Mark"] 
friend(x)
```

```python
'''
  Is this a triangle?
'''
def is_triangle(a, b, c):
  if a==0 or b==0 or c==0:
    return False
  elif a+b>c and a+c>b and b+c>a:
    return True
  else: 
    return False


#a+b>c
#a+c>b
#b+c>a
is_triangle(0,1,2)
```

```python
'''
  Who likes it?
  []                                -->  "no one likes this"
  ["Peter"]                         -->  "Peter likes this"
  ["Jacob", "Alex"]                 -->  "Jacob and Alex like this"
  ["Max", "John", "Mark"]           -->  "Max, John and Mark like this"
  ["Alex", "Jacob", "Mark", "Max"]  -->  "Alex, Jacob and 2 others like this"

'''

def likes(names):
  if len(names)==0:
    return print("no one likes this")
  elif len(names)==1:  
    return print("{} likes this".format(names[0]))
  elif len(names)==2:
    return print("{} and {} like this".format(names[0],names[1]))
  elif len(names)==3:
    return print("{}, {} and {} like this".format(names[0],names[1],names[2]))
  else:
    numbers = len(names) - 2
    return print("{}, {} and {} others like this".format(names[0],names[1],numbers))            

likes([])                                
likes(["Peter"])
likes(["Jacob", "Alex"])                 
likes(["Max", "John", "Mark"])           
likes(["Alex", "Jacob", "Mark", "Max"])
```

```python
'''
  Valid Braces?
  "(){}[]"   =>  True
  "([{}])"   =>  True
  "(}"       =>  False
  "[(])"     =>  False
  "[({})](]" =>  False
'''
import re
def validation(string):
  op = 0
  cp = 0
  ob = 0
  cb = 0
  ok = 0
  ck = 0
  for char in string:
    if char == '(':
      op += 1
    if char == ')':
      cp += 1
    if char == '[':
      ob += 1
    if char == ']':
      cb += 1
    if char == '{':
      ok += 1
    if char == '}':
      ck += 1      
  if (op == cp and ob == cb and ok==ck):
    return True
  else:
    return False                       

def validBraces(string):
    p = bool(re.search('(\[\)|\[\}|\(\]|\(\}|\{\)|\{\]|\)\()',string))
    if p == True:
      return False
    else:
      b = validation()
      if b:
        return True
      else:
        return False
    #[) #({
    #[} #)(
    #(]
    #(}
    #{)
    #{]


validBraces("([{}])")
```
