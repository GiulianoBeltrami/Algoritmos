```python
class ListNode(object):
     def __init__(self, val=None, next=None):
         self.val = val
         self.next = next

     def insert(self,value):
       newNode   = ListNode(value)
       self.next = newNode
```

```python
'''
  Course Schedule II
  There are a total of numCourses courses you have to take, labeled from 0 to numCourses - 
  1. You are given an array prerequisites where prerequisites[i] = [ai, bi] indicates that you must take course bi first if you want to take course ai.
  Return the ordering of courses you should take to finish all courses. If there are many valid answers, return any of them. If it is impossible to finish all courses, return an empty array
'''

import collections

def findOrder(numCourses: int ,prerequisites: list) -> list:
  '''
    numCourses : inteiro total de cursos que eu preciso ter
      range: 0 -> numCourses - 1

    prerequisites: lista de listas : [[0,1],[1,2]]
      formato: [A,B] , onde precisa de B para fazer A

    retorno: ordem dos cursos para finalizar todos
    EX: [[1,0],[2,0],[3,1],[3,2]] -> [0,2,1,3]
  '''
  #Algoritmo : Depth First Traversal (or Search)
  
  preq = {i:set() for i in range(numCourses)}
  graph =collections.defaultdict(set)

  for i,j in prerequisites:
		# preq armazena os requerimentos -> Depois:{Antes} 1:{0} -> precisa fazer 0 para fazer 1
    preq[i].add(j)
		# graph armazena os vizinhos     -> 0:{1,2} -> 0 tem linha para 1 e 2
    graph[j].add(i)
  print("Requerimentos antes: ",preq)
  print("Vizinhos: ",graph)

  #elemento parecido com uma lista mas com append e pops mais rápidos
  noPreq = collections.deque([])

  #adicionar a lista qual curso não possui pre requisito
  for x,y in preq.items():
    if len(y) == 0:
      noPreq.append(x)
  print("Matéria sem prerequisitos: ",noPreq)

  #track de quais cursos já foram feitos
  taken = []
  

  while noPreq:
    course = noPreq.popleft()
    taken.append(course)
    print("Cursos já feitos: ",taken)
    if len(taken) == numCourses:
      return taken

    for item in graph[course]:
      print("Item: ",item)
      print("Preq[item]: ", preq[item])
      print("Requerimentos: ",preq)
      preq[item].remove(course)
      if not preq[item]:
        noPreq.append(item)
  return []

numCourses = 4
prerequisites =[[1,0],[2,0],[3,1],[3,2]]
findOrder(numCourses,prerequisites)
```
